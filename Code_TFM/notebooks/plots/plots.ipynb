```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ajuste de ruta relativa desde CodeTFM/notebooks/plots/ hacia CodeTFM/
path = "../../estadisticas_comparativas_RAG.csv"
df = pd.read_csv(path)

# 1. Gráfico de métricas de calidad
metrics = ['faithfulness', 'answer_relevancy', 'context_precision', 'context_recall']
df_melted = df.melt(id_vars='rag_type', value_vars=metrics, var_name='Metric', value_name='Score')

plt.figure(figsize=(12, 6))
sns.barplot(data=df_melted, x='Metric', y='Score', hue='rag_type')
plt.title('Comparativa de Métricas de Calidad RAG')
plt.ylim(0, 1.1)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# 2. Gráfico de Latencia
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='rag_type', y='avg_latency_s', palette='viridis')
plt.title('Latencia Promedio por Tipo de RAG')
plt.ylabel('Latencia (segundos)')
plt.xlabel('Tipo de RAG')
plt.tight_layout()
plt.show()
```
