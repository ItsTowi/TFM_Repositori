```python
import os
import sys
from dotenv import load_dotenv

sys.path.append(os.path.abspath('..'))
load_dotenv(os.path.join('..', '.env'))
```

```python
from src.data_loader import load_and_split_text
from src.baselines.traditional_rag import TraditionalRAG
```

```python
ARCHIVO_ACTUAL = "../data/raw/book/christmas_carol_first_chapter.txt"
CARPETA_DB = "../chroma_db_experimentos" 
```

```python
rag = TraditionalRAG(persist_directory=CARPETA_DB)

if not os.path.exists(CARPETA_DB):
    splits = load_and_split_text(ARCHIVO_ACTUAL)
    print(f"Se van a procesar {len(splits)} fragmentos.")
    rag.index_documents(splits)
else:
    rag.load_existing_index()
```
