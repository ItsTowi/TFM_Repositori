```python
import sys
import os

repo_path = os.path.abspath("../../../LiteRAG")
if repo_path not in sys.path:
    sys.path.append(repo_path)

from literag import LiteRAG, LiteRAGConfig

print("✅ LiteRAG importado correctamente desde:", repo_path)
```

```python
# Si el yaml está en la raíz del repo LiteRAG, ajusta la ruta:
config_path = "../../../LiteRAG/literag_config.yaml" 

print("⚙️ Loading Configuration...")
config = LiteRAGConfig.from_yaml(config_path)
print("✅ Configuración cargada.")
```

```python
print("🚀 Initializing LiteRAG Engine (Loading Graph & LanceDB)...")
engine = LiteRAG(config)

# force_reload=False usa la caché para que las siguientes veces sea instantáneo
engine.initialize(force_reload=False)
print("✅ Engine listo.")
```

```python
# Tu nueva pregunta personalizada
query = "Tell me about Linux Kernel Networking"

print(f"🔍 Executing Query: '{query}'")
print("-" * 64)

# Ejecutamos la consulta (expand_query ayuda a encontrar sinónimos en el grafo)
result = await engine.aquery(query, expand_query=True)

if not result.success:
    print(f"❌ Error occurred: {result.error}")
else:
    print("✅ Consulta completada.")
```

```python
print("\n✅ ANSWER")
print("=" * 64)
print(result.answer)

print("\n📊 TELEMETRY & PERFORMANCE METRICS")
print("=" * 64)
print(f"⏱️  Latency:            {result.latency_ms / 1000:<4.2f} seconds")
print(f"⚓  Anchors Found:      {len(result.anchors):<4} starting nodes")
print(f"🕸️  Entities Explored:   {result.total_entities_explored:<4} nodes traversed")
print(f"🧠  Entities in Prompt: {result.entities_in_context:<4} highly-relevant nodes kept")
print(f"🪙  Context Tokens Used: {result.tokens_used:<4}")
print(f"📈  Token Breakdown:    {result.prompt_tokens} (Prompt) / {result.completion_tokens} (Completion)")
print("=" * 64)
```
